```python
import numpy as np
import pandas as pd
```

```python
casos = pd.read_csv("CasosActivosPorComuna.csv", index_col="Region")
casos.head()
```

```python
casos = casos.drop(columns=["Codigo region", "Comuna", "Codigo comuna", "Poblacion"])
casos = casos.groupby("Region").sum()
casos = casos.transpose()
casos = casos.loc['2020-07-01':'2021-07-01']
casos.head()
```

```python
paso = pd.read_csv("paso_a_paso.csv", index_col="region_residencia")
paso.head()
```

```python
paso = paso.drop(columns=["codigo_comuna", "comuna_residencia", "zona", "codigo_region"])
paso = paso.groupby("region_residencia").max().transpose()
paso.head()
```

```python
ventas = pd.read_excel("ventas_por_region.xlsx", index_col="Mes y aÃ±o")
ventas.drop('Total', axis=1)
ventas.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
fig, ax = plt.subplots(tight_layout=True,figsize=(12,7))
casos.plot(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(tight_layout=True,figsize=(12,7))
fmt_month = mdates.MonthLocator()
ax.xaxis.set_major_locator(fmt_month)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.set_xlabel('Fecha')
ax.set_ylabel('Etapa Plan Paso a Paso')
paso.plot(ax=ax)
plt.show()
fig.autofmt_xdate()
```

```python
fig, ax = plt.subplots(tight_layout=True,figsize=(12,7))
ax.set_xlabel('Fecha')
ax.set_ylabel('Millones de Pesos')
ventas.loc['2020-07-01':'2021-07-01'].plot(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(tight_layout=True,figsize=(12,7))
ax.set_xlabel('Fecha')
ax.set_ylabel('Millones de Pesos')
ventas["Total"].loc['2020-07-01':'2021-05-01'].plot(ax=ax)
plt.show()
```
